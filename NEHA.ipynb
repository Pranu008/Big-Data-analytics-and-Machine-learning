```python
# ISLP is a package used for examples from 'An Introduction to Statistical Learning'
from ISLP import load_data
from ISLP.models import (ModelSpec as MS, summarize)


from ISLP import confusion_table
from ISLP.models import contrast

# Importing ML models from scikit-learn
from sklearn.discriminant_analysis import (
LinearDiscriminantAnalysis as LDA,
QuadraticDiscriminantAnalysis as QDA
)
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# SECTION 2: LOADING THE SMARKET DATA

# The Smarket dataset contains stock market daily % returns.
# We use it to practice logistic regression.


Smarket = load_data('Smarket')
print(Smarket) # Display dataset
```

```python
# Viewing column names
print(Smarket.columns)
```

```python
# Checking correlations among numeric variables
print(Smarket[["Lag1", "Lag2", "Lag3", "Lag4", "Lag5", "Volume", "Today"]].corr())
```

```python


# Plotting volume over time
Smarket.plot(y='Volume');
```

```python
 ##SECTION 3: FITTING A LOGISTIC REGRESSION MODEL
 # We create a design matrix using all predictors except Today, Direction, Year.
# Then we fit a logistic regression using GLM with a binomial family.




allvars = Smarket.columns.drop(['Today', 'Direction', 'Year'])
design = MS(allvars)
X = design.fit_transform(Smarket) # Creates model matrix
y = (Smarket.Direction == 'Up') # Convert Up/Down into boolean for GLM


# Fitting the logistic model
glm = sm.GLM(y, X, family=sm.families.Binomial())
results = glm.fit()


# Summary of model
summarize(results)
```

```python
# Checking coefficients and p-values
print(results.params)
print(results.pvalues)
```

```python
# Predicting probabilities for the first 10 rows
probs = results.predict()
print(probs[:10])
```

```python
# Converting probabilities to predicted class labels
labels = np.array(['Down'] * len(probs))
labels[probs > 0.5] = "Up"
print(labels)
```

```python
# Confusion matrix to evaluate model performance
print(confusion_table(labels, Smarket.Direction))
```

```python
# Computing accuracy
print((507+145)/1250, np.mean(labels == Smarket.Direction))
```

```python
# SECTION 4: TRAINING/TEST SPLIT USING YEAR

# Here I split the data manually using the Year column,
# training on years < 2005 and testing on the rest.

train = (Smarket.Year < 2005)
Smarket_train = Smarket.loc[train]
Smarket_test = Smarket.loc[~train]


print(Smarket_test.shape)


X_train = X.loc[train]
X_test = X.loc[~train]
y_train = y.loc[train]
y_test = y.loc[~train]
```

```python
# Fitting model again on training data
glm_train = sm.GLM(y_train, X_train, family=sm.families.Binomial())
results = glm_train.fit()
probs = results.predict(exog=X_test)


D = Smarket.Direction
L_train, L_test = D.loc[train], D.loc[~train]


labels = np.array(['Down'] * len(probs))
labels[probs > 0.5] = 'Up'
print(confusion_table(labels, L_test))
```

```python


# Accuracy and error rate
print(np.mean(labels == L_test), np.mean(labels != L_test))
```

```python
# SECTION 5: FITTING THE MODEL WITH ONLY LAG1 & LAG2

# We now reduce the feature set to Lag1 and Lag2
# as suggested in the ISLR book.

model = MS(['Lag1', 'Lag2']).fit(Smarket)
X = model.transform(Smarket)
X_train = X.loc[train]
X_test = X.loc[~train]


glm_train = sm.GLM(y_train, X_train, family=sm.families.Binomial())
results = glm_train.fit()
probs = results.predict(exog=X_test)
labels = np.array(['Down'] * len(probs))
labels[probs > 0.5] = 'Up'


print(confusion_table(labels, L_test))
print((35+106)/252, 106/(106+76))
```

```python
# Example prediction for new values of Lag1 and Lag2
newdata = pd.DataFrame({'Lag1':[1.2, 1.5], 'Lag2':[1.1,-0.8]})
newX = model.transform(newdata)
print(results.predict(newX))
```

```python
# SECTION 6: PCA ON USArrests DATA
# In this section we switch to PCA to understand dimensionality reduction.


from statsmodels.datasets import get_rdataset
from sklearn.decomposition import PCA


USArrests = get_rdataset('USArrests').data
print(USArrests)
print(USArrests.columns)
print(USArrests.mean())
print(USArrests.var())
```

```python
# Scaling the data before PCA
scaler = StandardScaler(with_std=True, with_mean=True)
USArrests_scaled = scaler.fit_transform(USArrests)
```

```python
# Initializing PCA
test_pca = PCA()
test_pca.fit(USArrests_scaled)


scores = test_pca.transform(USArrests_scaled)
print(test_pca.components_)
```

```python
# Plotting the first two PCs
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(scores[:,0], scores[:,1])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
```

```python
# Adding arrows showing variable directions
for k in range(test_pca.components_.shape[1]):
    ax.arrow(0, 0, test_pca.components_[0,k], test_pca.components_[1,k])
    ax.text(test_pca.components_[0,k], test_pca.components_[1,k], USArrests.columns[k])
```

```python
# SECTION 7: KMEANS CLUSTERING

# Here I apply KMeans clustering to the USArrests dataset.

from sklearn.cluster import KMeans


X = USArrests[['Murder', 'Assault']]
kmeans = KMeans(n_clusters=3, random_state=42)
USArrests['Cluster'] = kmeans.fit_predict(X)


print(USArrests.head())


# Scatter plot of clusters
plt.scatter(USArrests['Murder'], USArrests['Assault'], c=USArrests['Cluster'], s=80)
plt.xlabel("Murder")
plt.ylabel("Assault")
plt.title("K-Means Clustering on USArrests")
plt.show()
```
